- Make a figure with subplots (matrix subplots)
- For some Distributions, plot the graphs of a PDF, CDF, generate random points, and plot some histogram in subplots. Add a title for each subplot, add some texts to plots
- Change the figure layout by using `tight_layout`
- Save the figure in .pdf, .png formats

```python
from scipy import stats
```

```python
x = np.linspace(-4, 4, 100)
y1 = stats.norm.pdf(x)
y2 = stats.norm.cdf(x)
r = stats.norm.rvs(size=200)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))
fig.tight_layout()                                      # fig.tight_layout(pad=3) distance between subplots
ax[0, 0].plot(x, y1)
ax[0, 0].set_title('PDF of Normal')
ax[0, 1].plot(x, y2)
ax[0, 1].set_title('CDF of Normal')
ax[1, 0].hist(r)
ax[1, 0].set_title('Histogram of random samples')

fig.savefig('normal.pdf')
```

```python
fig, ax = plt.subplots(5, 5, figsize=(12, 12))
for i in range(5):
    for j in range(5):
        m = x[5 * i + j].reshape(8, 8)
        ax[i, j].imshow(m)
        ax[i, j].set_title(y[5 * i + j])
        
fig.tight_layout(pad=2)        
```
