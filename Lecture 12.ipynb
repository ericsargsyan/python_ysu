- Calculate the approximate value of the 
$$
\int_0^1 \sin(x) dx,
$$
using the `midpoint rectangles` rule and `trapezoids` tule

```python
def SinInt_Midpoint(a, b, n=10):
    x = np.linspace(a, b, n + 1)
    h = (b - a) / n
    t = np.arange(a + h / 2, b, h)
    Sum = np.sum(np.sin(t) * h)
    
    true_value = np.cos(a) - np.cos(b)
    error = np.abs(Sum - true_value)
    return Sum, error
```

```python
SinInt_Midpoint(a=0, b=2, n=10000)
```
