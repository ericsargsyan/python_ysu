# Working with Polynomials

- Construct the Best fitting 1st, 2nd and 3rd order polynomials for the data
$$
(0,1); (2, -1); (3, 0); (-1, 4)
$$
Visualise the result.

```python
x = np.array([1, 2])
y = np.array([1, 2])
```

```python
p1 = np.polyfit(x, y, deg=1)
p1
```

```python
x1 = np.linspace(-1, 3, 100)
y1 = np.polyval(p1, x1)
```

```python
plt.plot(x1, y1)
plt.scatter(x,y);
```

```python
x = np.array([0, 2, 3, -1])
y = np.array([1, -1, 0, 4])
```

```python
plt.scatter(x, y)
```

```python
p = np.polyfit(x, y, deg=1)
```

```python
x1 = np.linspace(-1, 3, 100)
y1 = np.polyval(p, x1)
```

```python
plt.scatter(x, y)
plt.plot(x1, y1, c='red')   # regression line
```

- Construct the Least Squares Polynomial Fitting some data.Visualize the dataset and the fitting line

```python
x = 2 * np.random.rand(200) - 1
```

```python
y = 1.3 * x - 1 + 0.5 * np.random.randn(200)
```

```python
plt.scatter(x, y)
```

```python
p = np.polyfit(x, y, deg=1)
p
```

```python
x1 = np.linspace(-1, 1, 200)
y1 = np.polyval(p, x1)
```

```python
plt.scatter(x, y)
plt.plot(x1, y1, c='red');
```

```python
x = 2 * np.random.rand(200) - 1
```

```python
y = 1.3 * x ** 2 - 1 + 0.5 * np.random.randn(200)
```

```python
plt.scatter(x, y);
```

```python
p = np.polyfit(x, y, deg=2)
p
```

```python
x1 = np.linspace(-1, 1, 200)
y1 = np.polyval(p, x1)
```

```python
plt.scatter(x, y)
plt.plot(x1, y1, c='red');
```
