- Plot the graph of $f(x) = 0.2\cdot x^2\cdot \sin\dfrac{1}{x}$ over $[0.001, 0.1]$. Make it dashed and green

- Add the graphs of $\pm 0.2\cdot x^2$, in red solid

- Add the maximum points to the graph, in blue

- Add a title "The graph of $x^2\sin\dfrac{1}{x}$"

- Add X and Y axis Labels

- Add a Legend

```python
x = np.linspace(0.001, 0.1, 300)
y = 0.2 * (x ** 2 * np.sin(1 / x))
z = 0.2 * (x ** 2)
```

```python
x0 = np.array([1 / (np.pi / 2 + 2 * np.pi * k) for k in range(2, 7)])
y0 = 0.2 * (x0 ** 2)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, y, linestyle="--", c='green', label="plot of f(x)") # same as 'g--'
plt.plot(x, z, 'r-', lw = 2, label="Plot of x^2")
plt.plot(x, -z, 'r-', lw = 2, label="Plot of -x^2")
plt.scatter(x0, y0, c='black', marker='x', s=60)
plt.title("Graph of " + r'$\sigma(x) 0.2\cdot x^2\cdot \sin\dfrac{1}{x}$')
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend();
```
