- Explore the `scipy.stats` package, [link](https://docs.scipy.org/doc/scipy/reference/stats.html)

```python
from scipy import stats
```

- Draw the PDF and CDF of the $\mathcal{N}(0,1)$ Distribution

```python
normal = stats.norm(loc=0, scale=1)
```

```python
normal.rvs(size=10)
```

```python
normal.pdf(0)   # Խտության ֆունկցիան
```

```python
normal.cdf(0)  # Բաշխման ֆունկցիան
```

```python
x = np.linspace(-4, 4, 200)
```

```python
y = normal.pdf(x)
```

```python
plt.plot(x, y)
```

```python
z = normal.cdf(x)
```

```python
plt.plot(x, z)
```

- Generate 1000 random numbers from the $Exp(2)$ Distribution, plot the Density Histogram of that Numbers, and add over that the PDF of the theoretical Distribution. Do the same with the $\mathcal{N}(2,3)$ Distribution

```python
exp_dist = stats.expon(1 / 2)           # 1 / lam
```

```python
rsample = exp_dist.rvs(1000)
rsample[:10]
```

```python
x = np.linspace(0, 9, 200)
y = exp_dist.pdf(x)
```

```python
plt.hist(rsample, density=True, bins=20)
plt.plot(x, y)
```

```python
normal = stats.norm(loc=2, scale=3)
rnum = normal.rvs(1000)
```

```python
x = np.linspace(-8, 13, 200)
y = normal.pdf(x)
```

```python
plt.hist(rnum, bins=30, density=True)
plt.plot(x, y)
```
