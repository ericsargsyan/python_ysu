- For the $\sin$ function, construct 0, 1, 2, 3 order Taylor Polynomials at the point 0 and draw all that functions on the graph

$$
\sin x = \sum_{n=0}^\infty (-1)^n\cdot \dfrac{x^{2n+1}}{(2n+1)!} = x - \frac{x^3}{3!}+\frac{x^5}{5!}-...
$$

$$
T_0(x) = 0
$$
$$
T_1(x) = x
$$
$$
T_2(x) = x
$$
$$
T_3(x) = x-\frac{x^3}{3!}
$$

```python
def Sin0Taylor(x, n):
    if n == 0:
        return 0
    elif n == 1:
        return x
    
    term = x
    Sum = term
    
    for i in range(3,n+1,2):
        term = - term * x**2/((i-1)*(i))
        Sum += term
            
    return Sum
```

```python
Sin0Taylor = np.vectorize(Sin0Taylor)
```

```python
Sin0Taylor(np.array([2,3]), 5)
```

```python
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-np.pi,np.pi, 100)
y = np.sin(t)
```

```python
z = Sin0Taylor(t, 8)
```

```python
z
```

```python
plt.plot(t,y)
plt.plot(t,z)
```
